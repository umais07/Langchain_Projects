# Imports

```python
import os
from dotenv import load_dotenv

load_dotenv()
groq_key = os.getenv("groq_key")

from langchain_core.prompts import ChatPromptTemplate,FewShotPromptTemplate, PromptTemplate
from langchain_groq import ChatGroq
```

```python
template = """i want you to act as a acting financial advisor for people. 
In an easy way, explain the basics of the {financial_concept} """
```

```python
prompt = ChatPromptTemplate.from_template(template)

#Usage 
prompt.format(financial_concept = 'Income tax')
```

```python
llm = ChatGroq(
    temperature = 0.8,
    api_key=groq_key,
    model="meta-llama/llama-4-scout-17b-16e-instruct",
    max_tokens=1024,
)
```

```python
chain  = prompt | llm
```

```python
result = chain.invoke({"financial_concept": "income tax"})
```

```python
result
```

# Few Shot Antonym

```python
examples = [
    {'word':'happy','antonym':'sad'},
    {'word':'tall','antonym':'short'},
]

example_formatter_template = """word: {word}, Antonym: {antonym}"""

third_prompt = PromptTemplate(
    input_variables=["word", "antonym"],
    template=example_formatter_template
)

third_prompt.format(word = '"Tall"', antonym = '"short"')
```

```python
# Giving the example to the llm
few_shot = FewShotPromptTemplate(
    examples=examples,
    example_prompt=third_prompt,
    prefix="Give the antonym of the following word.\n",
    suffix="Word: {input}\nAntonym:",
    input_variables=["input"]
)

chain3 = few_shot | llm 
result = chain3.invoke({"input": "cold"})
```
